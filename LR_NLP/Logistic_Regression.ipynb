```python

import random

# Open the original training file
with open('train_processed.txt', 'r') as f:
    data = f.read()

# Split the data by empty lines to get individual sentences
sentences = data.strip().split('\n\n')

# Shuffle the sentences randomly
random.shuffle(sentences)

# Calculate the index to split the sentences
split_idx = int(len(sentences) * 0.8)

# Divide the sentences into train and dev sets
train_sentences = sentences[:split_idx]
dev_sentences = sentences[split_idx:]

# Save the train and dev sets to separate files
with open('train_set.txt', 'w') as f:
    f.write('\n\n'.join(train_sentences))

with open('dev_set.txt', 'w') as f:
    f.write('\n\n'.join(dev_sentences))
```

```python
import numpy as np
from gensim.models import KeyedVectors
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load pre-trained word vectors
word_vectors = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/word2vec/trunk/vectors.bin', binary=True)

# Load train data
train_data = []
with open('/content/data/train_set.txt') as f:
    for line in f:
        tokens = line.strip().split()
        sentence = [word for word in tokens[::2]]
        pos_tags = [pos for pos in tokens[1::2]]
        train_data.append((sentence, pos_tags))

# Load dev data
dev_data = []
with open('/content/data/dev_set.txt') as f:
    for line in f:
        tokens = line.strip().split()
        sentence = [word for word in tokens[::2]]
        pos_tags = [pos for pos in tokens[1::2]]
        dev_data.append((sentence, pos_tags))

# Get sentence vectors
def get_sentence_vector(sentence, model):
    # Check if the input sentence is not empty
    if sentence:
        # Extract words from the input sentence
        words = [word for word in sentence]
        # Get word vectors for all words in the sentence
        vectors = [model[word] if word in model else np.zeros(model.vector_size) for word in words]
        # Compute the sentence vector as the mean of the word vectors
        sentence_vector = np.mean(vectors, axis=0)
    else:
        # If the input sentence is empty, return a vector of zeros
        sentence_vector = np.zeros(model.vector_size)
    return sentence_vector


# Convert training data to feature matrix
X_train = np.vstack([get_sentence_vector(sentence, word_vectors) for sentence, pos_tags in train_data])
y_train = np.array([1 if pos_tags.count('NN') >= pos_tags.count('VB') else 0 for sentence, pos_tags in train_data])

# Train logistic regression classifier
clf = LogisticRegression(random_state=0).fit(X_train, y_train)

# Convert dev data to feature matrix
X_dev = np.vstack([get_sentence_vector(sentence, word_vectors) for sentence, pos_tags in dev_data])
y_dev = np.array([1 if pos_tags.count('NN') >= pos_tags.count('VB') else 0 for sentence, pos_tags in dev_data])

# Evaluate classifier on dev data
y_pred = clf.predict(X_dev)
accuracy = accuracy_score(y_dev, y_pred)
print("Accuracy on dev set:", accuracy)
```
